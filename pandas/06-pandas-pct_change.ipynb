## Percent Change 函数

pandas 的 percent change 函数用来计算当前值与指定周期(periods 默认为1)的比值的。 

- 先以一年内的营业额，利润作为例子

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'income': [10, 20, 30, 40, 50, 40, 50, 20, 20, 30, 20, 50],
    'profits':[5, 15, 25, 30, 40, 20, 20, 5, 15, 15, 20, 50]
    })
```

```python
df
```

```python
## 计算本月与上月，营业额及利润的环比增长率。即拿 （本月 - 上月）/ 上月
df.pct_change() # 默认 periods = 1
```

```python
## 计算，季度的同比增长率。 即 (本月 - 上季度对应月份)/上长度对应月份
print(df)
df.pct_change(periods=3) 
```

```python
## 默认 axis=0, 那么计算实收率的话，就要按 columns 来算了，所以指定 axis=1
# 计算实收率
print(df)
pct = df.pct_change(axis=1)
print(pct)
df['ratio'] = pct['profits'].apply(lambda x: (1 + x) * 100)
df
```
