```python
'''
使用 python 内置 csv 模块来读取 csv 文件。
'''
import matplotlib.pyplot as plt
import csv

x = []
y = []
with open('example.csv') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
plt.plot(x,y)
plt.show()
```

```python
'''
使用 numpy 来加载文本文件，速度更快更方便。
'''
import numpy as np
import matplotlib.pyplot as plt

rows = np.loadtxt('example.csv', delimiter = ',', unpack = False) # unpack 指的是是否解析，为 False 时会按行返回。
for row in rows:
    print(row)

x, y = np.loadtxt('example.csv', delimiter = ',', unpack = True)
plt.scatter(x, y)
plt.show()
```
