## Plotting Data

```octave
clf(); % clear figure
t = [0:0.01:0.98];
y1 = sin(2 * pi * 4 * t);
y2 = cos(2 * pi * 4 * t);
plot(t, y1, 'r');
hold on; % Retain plot data and settings so that subsequent plot commands are displayed on a single graph.
plot(t, y2);
xlabel('time')
ylabel('value')
legend('sin', 'cos')
title('My Plot')
print -dpng myPlog.png
```

```octave
close all % close all plot figure
```

```octave
figure(1); plot(t, y1, 'r')
figure(2); plot(t, y2)
```

```octave
close all;
subplot(1, 2, 1) % Divide plot into a  1 x 2 grid, and access 1st block.
plot(t, y1, 'r')
axis([-0.2, 1])
subplot(1, 2, 2) % access 2nd block
plot(t, y1, 'b')
axis([-0.5, 1])
```
