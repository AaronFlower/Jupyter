```python
import numpy as np
import matplotlib.pyplot as plt

# 不做数据归一化感觉计算值会比较大的。

def loadData(file):
    data = np.loadtxt(file, delimiter=',')
    features = np.insert(data[:,0:2], 0, 1.0, axis = 1) # np.insert
    prices = data[:,2][:, np.newaxis] # np.newaxis
    return features, prices

def LSE_weights(data, prices):
    X = np.mat(data)
    y = np.mat(prices)
    return (X.T * X).I * X.T * y


def plotPriceRegression():
    features, prices = loadData('data/portland-housing-price.txt')
    weights = LSE_weights(features, prices)
    print(weights)
    plt.figure(figsize=(15, 4), dpi=100)
    plt.subplot(121)
    plt.xlabel('Size of house (X1)')
    plt.ylabel('House Price (Y)')
    x_1 = np.arange(500, 5000, 1)
    x_2 = -(weights[0] + weights[1] * x_1) / weights[2]
#     plt.scatter(features[:, 1], prices.flatten())
    print(np.squeeze(np.asarray(x_2)))
    plt.plot(x_1, np.squeeze(np.asarray(x_2)))
    plt.subplot(122)
    plt.xlabel('Number of Bedrooms (X2)')
    plt.ylabel('House Price (Y)')
    plt.scatter(features[:, 2], prices.flatten())
    
    plt.show()
    
if __name__ == '__main__':
    plotPriceRegression()
```
