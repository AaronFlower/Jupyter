## 概率与数理统计

TL;DR;

- 古典概型，样本空间有限，样本点等可能。
- 条件概率的样本空间会收缩变小。
- 离散型随机变量用**分布律**表示；
- 连续型随机变量用**概率密度函数**表示；
- 离散型随机变量和连续型随机变量都可以用**随机变量分布函数**统一描述；
- 随机变量函数的分布；（随机变量作为函数的自变量）

### 1. 随机事件

#### 1.1 基本概念

- 样本空间 $\Omega$: 试验的所有可能的结果组成的集合称为样本空间。
- 样本点：样本空间的元素，也就是随机试验的单个结果称为样本点。
- 事件：样本空间的任意一个子集称为随机事件，简称为事件。
- 基本事件：一个随机事件中只你包含一个试验结果，称之为基本事件。

Ex: 掷骰子，观察所掷的点数。

1. 其样本空间为 $\Omega = \{1, 2, 3, 4, 5, 6\}$
2. $A_1 = \{1\}, A_2 = \{2\},A_3 = \{3\},A_4 = \{4\},A_5 = \{5\},A_6 = \{6\}$ 是基本事件。
3. $ B = \{偶数点\}， C = \{奇数点\}， D = \{4 或 4 点以上\} $ 是事件

- $\Omega$ 为必然事件
- $\emptyset$ 为不可能事件



#### 1.1.2 事件的关系与运算

既然事件是一个集合，因此事件间的关系、运算及运算规则也就按集合的关系、运算及运算规则来处理。

1. 若 $A$ 发生时，$B$ 必然发生，则 $A  \subset B$. 若 $ A \subset B, B \subset A$ , 则 $A = B$.
2. 对于两上事件 $A$ 和 $B$, 定义一个新事件, $C = \{ A 或 B 发生\}$, 则 $ C = A \cup B $ 或 $C = A + B$.
3. 对于两上事件 $A$ 和 $B$, 定义一个新事件, $C = \{ A 和 B 都发生\}$, 则 $ C = A \cap B $ 或 $C = AB$. 若 $AB = \emptyset$ 则，$A$ 与 $B$ 为互斥事件，简称为 $A$ 与 $B$ 互斥。
4. 对于两个事件 $A$ 和 $B$，定义一个新事件, $C = {A 发生，B 不发生}$，则记为 $C = A - B$, 称为事件 $A$ 与 $B$ 的差。
    - 特别地：$\bar{A} = \Omega - A$, 称为 $A$ 的对立事件，或补事件。
    - $A \cup \bar{A} = \Omega$
    - $A \cap \bar{A} = \emptyset$

在事件的运算过程中，经常用到如下规则:

- 交换律: $A \cup B = B \cup A, A\cap B = B \cap A $.
- 结合律: $A \cup (B \cup C) = (A \cup B) \cup C, A (B C) = (A  B) C$
- 分配律: $A(B \cup C) = (AB) \cup (AC), A\cup(BC) = (A \cup B)(A \cup C)$
- 对偶律: $\overline{A \cup B} = \bar{A}\bar{B}, \overline {A \cap B} = \bar{A} \cap \bar{B}$ 
- 还有：$A - B = A\bar{B}, A = (AB)\cup(A\bar{B})$

### 1.2 事件的概率

Def. 设 $E$ 是随机试验，$\Omega$ 是它的样本空间，对于每个事件 $A$, 定义一个实数 $P(A)$ 与这对应。若函数 $P(\cdot)$ 满足条件：

1. 对于每个事件 A, 均有 $ P(A) \ge 0$

2. $P(\Omega) = 1$

3. 若 $A_1, A_2, ...$ 两两互斥，即对于任意 $i, j = 1, 2, ...; i \ne j, A_i A_j = \emptyset $ 均有：$ P(A_1 \cup A_2 \cup ...) = P(A_1) + P(A_2) + ... $

则称 $P(A)$ 为事件 A 的概率。这里 $P(\cdot)$ 函数的自变量是一个集合，因此它是一个集合函数。

概率有如下性质：

1. $P(\emptyset) = 0$, 即不可能事件的概率为零。
2. 若 $A_1, A_2, ... A_n$ 两两互斥，则$ P(A_1 \cup A_2 \cup ... A_n) = P(A_1) + P(A_2) + ... + P(A_n)$.
3. $\forall A, P(\bar{A}) = 1 - P(A) $
4. 若 $A\subset B$, 则 $P(B - A) = P(B) - P(A), P(B) \ge P(A)$
5. 若 $\forall A, B, P(A \cup B) = P(A) + P(B) - P(AB)$

$$
    P(A_1 \cup A_2 \cup A_3) = P(A_1) + P(A_2) + P(A_3) - P(A_1A_2) - P(A_2A_3) - P(A_1A_3) + P(A_1A_2A_3)
$$





### 1.3 古典概率模型

若实验 E 的结果只有有限种，且每种结果发生的可能性相同，则称这样的试验模型为**等可能概率模型** 或 **古典概率模型** 简称为**等可能概型**或**古典概型**。

#### 1.3.1 古典概型的计算问题

设试验 E 的样本空间为: $\Omega = \{\omega_1, \omega_2, \ldots, \omega_n\}$, 且在试验中每个基本事件发生的可能性相同，即: $P(\omega_1) = P(\omega_2) = \ldots = P(\omega_n) $。

由基本事件两两互斥，且 $\{\omega_1\} \cup \{\omega_2\} \cup \ldots \cup\{ \omega_n\} = \Omega$ 得：

\begin{align}
    1 &= P(\Omega) \\
    &= P(\{\omega_1\} \cup \{\omega_2\} \cup \ldots \cup\{ \omega_n\}) \\
    &= P(\omega_1) + P(\omega_2) + \ldots + P(\omega_n) \\
    &= nP({\omega_i})
\end{align}

所以，$P(\omega_1) = P(\omega_2) = \ldots = P(\omega_n) = \frac{1}{n} $。

若事件 A 包含 k 个基本事件，则 

$$
P(A) = \sum_{i = 1}^{k}P({\omega_i}) = \frac{k}{n} = \frac{\#A}{\#\Omega}
$$

### 1.4 条件概率

$P(A|B), P_B{A}$, 读作事件 B 发生的条件下，事件 A 发生的概率。

Def. 设 $A, B$ 两个事件：

$$
P(A|B) = \frac{P(AB)}{P(B)}, P(B) > 0.
$$

称这为事件 B 发生的条件下，事件 A 发生的条件概率。

1. $P(A|B)\ge 0$
2. $P(\Omega|B) = 1$
3. $P(A_1 \cup \ A_2 \cup \ldots | B) = P(A_1|B) + P(A_2|B) + P(A_3|B) + \ldots$

**注意**：条件概率的样本空间改变了，收缩变小了。

#### 1.4.2 乘法公式

\begin{align}
    P(AB) &= P(B) P(A|B)\\
    P(AB) &= P(A) P(B|A)\\
    P(ABC) &= P(A)P(B|C)P(A|BC)
\end{align}

#### 1.4.3 全概率公式

为介绍全概率公式，我们先引入空间划分的概念。

**Def.** 设 $\Omega$ 为试验 E 的样本空间，$B_1, B_2, \ldots, B_n$ 为一组事件，若 $B_1, B_2, \ldots, B_n$ 两两互斥，且 $B_1 \cup B_2 \cup \ldots \cup B_n = \Omega $, 则称   $B_1, B_2, \ldots, B_n$ 为样本空间 $\Omega$ 的一个划分。

**定理**： 设 $\Omega$ 为试验 E 的样本空间，A 为一个事件，$B_1, B_2, \ldots, B_n$ 为样本空间 $\Omega$ 的一个划分， 且 $P(B_i) \gt 0, i = 1, 2, \ldots n$, 则有:

$$
    P(A) = \sum_{i = 1}^{n}P(B_i)P(A|B_i)
$$

上式称这为全概率公式。

在实际的问题中 $P(A)$ 不易直接求出，但却容易找到 $\Omega$ 的一个划分。

#### 1.4.4 贝叶斯公式


**定理**： 设 $\Omega$ 为试验 E 的样本空间，A 为一个事件，$B_1, B_2, \ldots, B_n$ 为样本空间 $\Omega$ 的一个划分， 
且 $P(A) > 0, P(B_i) \gt 0, i = 1, 2, \ldots n$, 则有:

$$
    P(B_i|A) = \frac{P(B_i)P(A|B_i)}{\sum_{j=1}^{n}P(B_j)P(A|B_j)}, i = 1, 2, ..., n
$$

上式称这为贝叶斯公式。


### 1.5 事件独立性

**Def**: 若 A 与 B 相互独立，如果等式 $$P(AB) = P(A)P(B)$$ 成立，则称事件 A 与 B 相互独立。

在实际的应用中，两个事件是否独立，我们常常不是根据上式来判断，而是根据这两个事件的发生是否相互影响来判断。

#### 2.3.4 随机变量分布函数 

离散型随机变量，我们用分布律刻划。连续型随机变量，我们用概率密度函数刻划。现在引入随机变量分布函数可以对两类变量进行描述。

定义：设 X 为随机变量，称函数  

$$
F(X) = P\{ X \le x \}, -\infty \lt +\infty
$$

为 X 的分布函数。 其有以下性质：

1. 对 $\forall a, b, a \lt b, 有 F(a) \le F(b)$, 单调非减性，并且 $P \{a < X \le b\} = F(b) - F(a)$。
2. $\forall x,  0 \le F(x) \le 1$, 有界性。 且:

$$
    F(-\infty) = \lim{x \to -infty} F(x) = 0 \\
    F(\infty) = \lim{x \to infty} F(x) = 1
$$

对于离散型 X, 其分布函数为:

$$
    F(X) = P\{X \le x\} = \sum_{x_k \le x} P\{x = x_k\}, 即 \\
    F(X) = \sum_{x_k \le x} p_k
$$

对于连续型 X, 设期概率密度函数为 $f(x)$, 则：

$$
F(x) = P\{ X \le x\} = \int_{-\infty}^{x}f(t)dt
$$

### 4. 数字特征

#### 4.1 期望

**Def 4.1.1**, 设离散型随机变量的概率分布为: 

$$
 P\{X = x_i\} = p_i, i = 1, 2, \ldots,
$$

若级数 $\sum{x_ip_i}$ 绝对收敛，即 $\sum{|{x_i}|p_i}$ 收敛，则称 $\sum{x_ip_i}$ 为随机变量 X 的期望，记为 E(X), 即：

$$
E(X) = \sum_i{x_ip_i}
$$

在 X 取值可列无穷个值时，级数 $\sum{x_ip_i}$ 绝对收敛可以保证 『级数的值不因级数各项次序的改变百变化』这样 E(X) 与 X 取的值与排序次序无关。

从公式上可以看出，期望就是随机变量 **X 的取值 $x_i$ 以它们的概率为权**的加权平均。所以期望也称为均值。

**Def 4.1.2**, 连续型随机变量 X 的概率密度函数为 $f(x)$, 若积分 $\int_{-\infty}^{+\infty}f(x)dx = E(X)$ 绝对收敛，则称 $E(X)$ 为随机变量 X 的期望，即

 $$
 E(X) = \int_{-\infty}^{+\infty}f(x)dx 
 $$
 
#### 4.1.4 期望的性质

1. c 是常数，$E(c) = c$
2. k 是常数，$E(kX) = kE(X)$
3. $E(X + Y) = E(X) + E(Y)$
4. 若 X, Y 相互独立，则 $E(XY) = E(X)E(Y) $


### 4.2 方差

方差是随机变量的另一个重要的数字特征，它刻画了随机变量取值在其中心位置的分散程度。

**Def 4.2.1**, 设 X 为一随机变量，如果 $E\{[X - E(X)]^2\}$ 存在，则称之为 X 的方差，记为 $Var(X)$, 即

\begin{align}
D(X) = Var(X) &= E\{[X - E(X)]^2\} \\
    &= E(X^2) - [E(X)]^2
\end{align}

并称 $\sqrt{Var(X)}$ 为 X 的标准差或均方差。 计算方差归结为计算两个期望 $E(X^2), E(X)$。

#### 4.2.2 方差的性质

1. 若 c 为常数，则 $Var(c) = 0, Var(X + c) = Var(C)$.
2. 若 k 为常数，则 $Var(kX) = k^2Var(X)$
3. 若 X 和 Y 相互独立，则 $Var(X + Y) = Var(X) + Var(Y)$
4. $ Y = \frac{X - E(X)}{\sqrt{Var(X)}}$ 为 X 的标准化随机变量，因为：

$$
E(Y) = \frac{E[X - E(X)]}{\sqrt{Var(X)}} = 0\\
Var(Y) = \frac{Var[X - E(X)]}{Var(X)} = 1
$$

### 4.3 协方差与相关系数

对于二维随机向量 $(X, Y)$, 除了它的分量 X, Y 的期望和方差之外，还有一些数字特征，用来刻画 X 与 Y 之间的相关程度。

#### 4.3.1 协方差

**定义4.3.1** 称 $E\{[X - E(X)][Y - E(Y)]\}$ 为 X 与 Y 的协方差，记为 Cov(X, Y), 即：

$$
Cov(X, Y) = E\{[X - E(X)][Y - E(Y)]\}
$$

从上式可以看出，当 $Cov(X, Y) > 0 $ 时两个随机变量的同向变化，否则有相反方向变化的趋势。

协方差具有下面四个性质：

1. $Cov(X, Y) = Cov(Y, X)$
2. 设 a,b,c,d 为常数，则 $ Cov(aX + b, cY + d) = a bCov(X, Y)$
3. $Cov(X_1 + X_2, Y) = Cov(X_1, Y) + Cov(X_2, Y)$
4. $Cov(X, Y) = E(XY) - E(X)E(Y), 当 X, Y 相互独立时，则 Cov(X, Y) = 0$

#### 4.2.3 相关系数

**定义 4.3.2**, 若 $Var(X) > 0, Var(Y) > 0$, 则称 $\frac{Cov(X, Y)}{\sqrt{Var(X)Var(Y)}}$ 为 X, Y 的相关系数，记作 $ \rho_{XY} $, 即：
$$
\rho_{XY} = \frac{Cov(X, Y)}{\sqrt{Var(X)Var(Y)}}
$$

上面的相关系数，也称之为皮尔逊相关系数。

相关系数 $\rho_{XY}$ 与协方差 $Cov(X, Y)$ 之间**相差一个倍数**，相关系数是标准尺度下的协方差。

协方差依赖于 X 与 Y 的度量单位，如果将 X 与 Y 都除以各自的标准差，这样就能更好的反映两着的关系，所以相关系数 $\rho_{XY}$ 不受所用的度量单位的影响。

另外，$X, Y$ 相互独立的充要条件是 $\rho_{XY} = 0$ .

**定理 4.3.2**: 

$$
    |\rho_{XY}| \le 1
$$
当且仅当 X 与 Y 之间有**线性关系**时等号成立， 即有线性的 $Y = aX + b$ 关系。

**定义 4.3.3** 若 X 与 Y 的相关系数  $\rho_{XY} = 0$, 则称 X 与 Y 互不相关。

当 $0 < |\rho_{XY}| < 1$ 时，
    - 若 $|\rho_{XY}|$ 较大，则表明 X, Y 之间线性关系的联系较**好**，越接近于 1，这种联系越好；
    - 若 $|\rho_{XY}|$ 较小，则表明 X, Y 之间线性关系的联系较**差**，越接近于 0，这种联系越差；
    - 当 $|\rho_{XY}| = 0 $ 时，X 与 Y i 不存在线性关系；
    - $\rho_{XY} > 0$ 称为正相关；
    - $\rho_{XY} < 0$ 称为负相关；

### 4.4 矩与协方差矩阵

#### 4.4.1

随机变量的另一个数字特征是矩，包括**原点矩**和**中心矩**, 它们在数理统计中有重要的应用。

**定义 4.4.1**，有 $k = 1, 2, \ldots $

- 对于随机变量 X ， 若 $E(X^k)$ 存在，则称它为 X 的 k 阶原点矩。
- 对于随机变量 X ， 若 $E\{[X - E(X)]^k\}$ 存在，则称它为 X 的 k 阶中心矩。

易知，期望是一阶原点矩，而方差是两阶中心矩。

#### 4.4.2 协方差矩阵

**定义 4.4.2**, 对于 n 维随机向量 $(X_1, X_2, \cdots, X_n)$, 记

$$
c_{ij} = E\{[X_i - E(X_i)][X_j - E(X_j)]\}, i,j = 1,2, \ldots, n \\
$$

称矩阵：


\begin{align}
C = c_{i,j} = 
 \begin{pmatrix}
  c_{1,1} & c_{1,2} & \cdots & c_{1,n} \\
  c_{2,1} & c_{2,2} & \cdots & c_{2,n} \\
  \vdots  & \vdots  & \ddots & \vdots  \\
  c_{n,1} & c_{n,2} & \cdots & c_{n,n} 
 \end{pmatrix}
\end{align}

为 $(X_1, X_2, \cdots, X_n)$ 的协方差矩阵。

显然，协方差矩阵是一个对称矩阵。

由于 n 维随机向量的分布在很多情况下不知道，或是过于复杂而不便使用，这时可以**使用其协方差矩阵**来帮助分析 n 维随机向量。如在 **PCA (Principal Component Analysis) 主成分析**来进行降维时就可以用到这个矩阵。

## 6. 总体，样本，极大似然估计

### 6. 1 总体与样本

**总体** $X$ , 即是我们研究问题所涉及到的对象的全体。

**样本** $X_1, X_2, ... X_n, ...$ , 是从总体随机抽取的一些样品。

**样本的二重性：** , 样本即是数，又是随机变量。

当样本  $X_1, X_2, ... X_n, ...$ 看成随机变量时，是相互独立且同分布的 （Independent and Indentically Distribued, IID） 并且与 总体 $X$ 的分布相同。

假设  $X$ 的概率密度是 $f(x)$, 则 $X_1, X_2, ... X_n$ 与 $ X $ IID, 有其联合概率密度函数：

\begin{equation*}
g(x_1, x_2, ..., x_n) = \prod_{i=1}^{n} f(x_i)
\end{equation*}

### 6.2 极大似然估计

设总体分布为 $f(x,\space \theta_1, ..., \theta_k)$,  $X_1, X_2, ... X_n$ 为从该总体抽出的样本， 因为  $X_1, X_2, ... X_n $ IID, 所以它们的联合概率密度函数为：

\begin{equation*}
L(x_1, x_2, ..., x_n;\space \theta_1, ..., \theta_k) = \prod_{i=1}^{n} f(x_i,\space \theta_1, ..., \theta_k)
\end{equation*}

我们把 $x_1, x_2, ..., x_n$ 看成固定的， 则 $L(x_1, x_2, ..., x_n;\space \theta_1, ..., \theta_k)$ 就是 $\theta_1, ..., \theta_k $ 的函数，这时我们称之为似然函数。

对 $\theta$ 的极大似然估计就归结为示 $L(\theta)$ 的最大值点。 一般极大依然函数都 log 来处理一下方便计算。

在 Logistic 回归中用到这个方法，求最大值点，当然要用梯度上升算法了。
