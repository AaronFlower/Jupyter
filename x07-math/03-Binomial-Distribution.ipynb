## Binomial Distribution 二项分布

二项分布就是重复 $n$ 次独立的伯努利实验(Bernoulli trial), 每次实验成功的概率为 $p$。 所以二项式分布有两个参数：

- $n$: $n$ 次独立的实验；
- $p$: 每次成功的概率 (true/false, 1/0), 相应的失败的概率为 $ q = 1 - p $

特别的，当 $n = 1$ 时，称之为二点分布或 Bernoulli 分布。

### 1.  PMF, Probability Mass Function

二项分布是一个离散性的分布，当 $n \in N, p \in [0, 1], X ~ B(n, p)$, 其概率质量函数(PMF, Probability Mass Function)为:

\begin{align}
f(k; n, p) &= Pr(k; n, p) \\
    &= Pr(X = k) \\
    &= C_{n}^{k}p^k(1 - p)^{n - k}
\end{align}

- $p^k$ 表示 $k$ 次成功的概率相乘；
- $(1 - p)^{n - k}$ 表示 $n - k$ 次失败的概率相乘；
- 在 n 次实验中，成功和失败的次序是无关紧要的，所以有 $C_{n}^{k}$ 种序列可取。

相应的对于连续性随机变量，我们用 PDF, Probability Density Function 概率密度函数来描述连续性随机变量。

### 2. CDF,  Cumulative distribution function 

CDF 累积分布函数，也叫分布函数，用来描述连续性随机变量。通过计算连续性随机变量函数的线下积分面积来描述。一般以 CDF 来标记：

$$
F_X(x) = P(X \le x)
$$

对于二项式分布的 CDF 可以写为:

$$
F(k; n, p) = Pr(X \le k) = \sum_{i = 0}^{i = k}C_n^k p ^ k (1 - p) ^ {n - k}
$$

### 3.  $ f(k; n, p) $

$ f(k; n, p) $ 这个函数就比较有趣了，这个函数是一个对称函数（偶函数）。 其中：

$$
f(k; n, p) = C_{n}^{k}p^k(1 - p)^{n - k} = C_{n}^{n-k}(1 - p)^{n - k} p^k = f(n-k; n, 1 - p)
$$

该函数是一个关于 $k$ 的函数，我们可以求出其最大值。通过下面的式子来求得：

\begin{align}
\frac{f(k+1; n, p)}{f(k; n, p)} &= \frac{ C_{n}^{k+1}p^{k+1}(1 - p)^{n - k-1}} { C_{n}^{k}p^k(1 - p)^{n - k}} \\
&= \frac{n-k}{k+1}\frac{p}{1 - p}
\end{align}

让上式与 1 做比较，可得到 $(n + 1)p - 1 \le k < (n + 1)p $, 该数称为众数(Mode)。

#### 5.2  Beta 分布

二项分布与 Beta 分布都对 n 次独立的 Bernoulli 实验的描述。只不过是二项分布使用的是离散型的概率质量函数（PMF)描述，而 Beta 分布使用的概率密度函数(PDF) 描述。

在数学上，当 $\alpha = k + 1, \beta = n - k + 1$ 时，Beta 分布与二项式分布成比例：

\begin{align}
Beta(p; \alpha; \beta) &= (n + 1) B(k; n; p) \\
&= (n + 1) C_{n}^{k}p^k(1 - p)^{n - k}
\end{align}

**Beta分布其实提供了一个后验概率的计算方法**，当 $\alpha, \beta$ 参数固定时，根据先验概率可以计算后验概率：

$$
P(p; \alpha, \beta) = \frac{1}{B(\alpha, \beta)} p^{\alpha - 1} (1 - p)^{\beta - 1}
$$

Beta 分布概率密度函数, 当 $\alpha, \beta$ 参数固定时：

\begin{align}
f(x;\alpha, \beta) &= constant \cdot x^{\alpha - 1} (1 - x)^{\beta - 1} \\
&= \frac{x^{\alpha - 1} (1 - x)^{\beta - 1}} {\int_0^1 u ^ {\alpha - 1} (1 - u)^{\beta - 1}d_u} \\
&= \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha)\Gamma(\beta)} x^{\alpha - 1} (1 - x)^{\beta - 1} \\
&= \frac{1}{B(\alpha, \beta)} x^{\alpha - 1} (1 - x)^{\beta - 1}
\end{align}

- $\Gamma(z)$ 是 Gamma 函数；
- $\frac{1}{B(\alpha, \beta)}$ 是一个常量，用来保证总概率为 1.
- 则随机变量 X 满足  X ~ $ Beta(\alpha, \beta)$
